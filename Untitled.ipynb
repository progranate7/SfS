```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# make shading data and lighting
L = [0.0, 0.0, 1.0]
sphere_shading = np.zeros((256, 256), dtype='float')
cx = 256/2
cy = 256/2
r = 110
for i in range(0, 256):
    for j in range(0, 256):
        if np.sqrt((i-cx)**2 + (j-cy)**2) > r:
            sphere_shading[j, i] = 1.0
        else:
            sphere_shading[j, i] = np.cos(np.arcsin(np.sqrt((i-cx)**2 + (j-cy)**2)/r))
        
#sphere_shading = cv2.GaussianBlur(sphere_shading, (5, 5), 3)
plt.imshow(sphere_shading)
plt.colorbar()
plt.show()

cv2.imwrite('dat/sphere_shading.png', (sphere_shading*255).astype('uint8'))

sphere_shading_noise = sphere_shading + np.random.normal(0, 0.1, (256, 256))
plt.imshow(sphere_shading_noise)
plt.colorbar()
plt.show()
```

```python
# shape from shading of linear approximation method
def R(height, light, shading):
    h, w = height.shape
    p = height[:(h-1), 1:w] - height[:(h-1), :(w-1)]
    q = height[1:h, :(w-1)] - height[:(h-1), :(w-1)]
    ret_val = np.zeros((h, w))
    ret_val[:(h-1), :(w-1)] = (L[2] + p*L[0] + q*L[1])/np.sqrt(1+p**2+q**2)
    ret_val[:, w-1] = ret_val[:, w-2]
    ret_val[h-1, :] = ret_val[h-2, :]
    ret_val[h-1, w-1] = ret_val[h-2, w-2]
    return ret_val
        
def f(shading, height, light):
    return shading - R(height, light, shading)
    
def diff_f(shading, height, light):
    h, w = height.shape
    ps = light[0] / light[2]
    qs = light[1] / light[2]
    p = height[:(h-1), 1:w] - height[:(h-1), :(w-1)]
    q = height[1:h, :(w-1)] - height[:(h-1), :(w-1)]
    pq1 = np.sqrt(p**2 + q**2 +1)
    psqs1 = np.sqrt(ps**2 + qs**2 + 1)
    M = np.zeros((h, w))
    M[:(h-1), :(w-1)] = (ps+qs)/(pq1*psqs1) - ((p+q)*(p*ps+q*qs+1))/(psqs1*(pq1**3))
    M[:, w-1] = M[:, w-2]
    M[h-1, :] = M[h-2, :]
    M[h-1, w-1] = M[h-2, w-2]
    return M    
    
def linear_SfS(shading, light):
    height = np.zeros_like(shading)
    M = diff_f(shading, height, light)
    K = 1/(0.001 + M)
    prev_height = np.copy(height)
    height = height - K*f(shading, height, light)

    error = []
    for i in range(0, 1000):
        M = diff_f(shading, height, light)
        # M[np.logical_and(M<0, np.abs(M) < 0.00001)] = -0.00001
        # M[np.logical_and(M>0, np.abs(M) < 0.00001)] = 0.00001
        # S = np.sqrt(np.mean((prev_height - height)**2))
        S = np.sqrt(np.mean(f(shading, height, light)**2))
        K = S*M / (0.001 + S * (M**2))
        # K = 1/M
        #print(M)
        print(S)
        #plt.imshow(M)
        #plt.colorbar()
        #plt.show()
        #plt.imshow(f(shading, height, light))
        #plt.colorbar()
        #plt.show()
        prev_height = np.copy(height)
        height = height - K*f(shading, height, light)
        #height = cv2.GaussianBlur(height, (3, 3), 0)
        #plt.imshow(height)
        #plt.colorbar()
        #plt.show()
        error.append(np.sum(f(shading,height,light)**2))
    
    plt.plot(np.array(error))
    plt.show()
    return height
```

```python
coin_shading = cv2.imread('dat/coin.png', cv2.IMREAD_GRAYSCALE)
coin_shading = coin_shading.astype('float')
coin_shading = cv2.GaussianBlur(coin_shading, (5, 5), 0)
coin_shading = (coin_shading - np.min(coin_shading)) / (np.max(coin_shading) - np.min(coin_shading))
tmp_height = linear_SfS(coin_shading, L)
plt.imshow(cv2.GaussianBlur(tmp_height, (7, 7), 0), cmap='jet')
plt.colorbar()
plt.show()

```
